## **Lib**

```python
import pandas as pd
import seaborn as sns
```

## **Import data**

```python
data = pd.read_csv("diamonds.csv")
data.head(10)
```

## **Replace data**

```python
data.drop(['Unnamed: 0'],axis=1,inplace=True)

new_colors = {"D": 7, "E": 6,"F": 5,"G": 4,"H": 3,"I": 2,"J": 1}
data["color"].replace(new_colors, inplace=True)

new_clarity = {"FL":10,  "IF": 9, "VVS1": 8, "VVS2": 7, "VS1": 6, "VS2": 5, "SI1": 4, "SI2": 3, "SI3": 2, "I1": 1}
data["clarity"].replace(new_clarity, inplace=True)

new_cut = {"Ideal": 5, "Premium": 4, "Very Good": 3, "Good": 2, "Fair": 1}
data["cut"] .replace(new_cut, inplace=True)

data.head(10)
```

**depth**

```python
data["depth"] = data["depth"].astype(int)
data["depth"] .replace(to_replace=range(58,63), value=3,inplace=True)
data["depth"] .replace(to_replace=range(58,64), value=2,inplace=True)
data["depth"] .replace(to_replace=range(10,82), value=1,inplace=True)
```

**Table**

```python
data["table"] = data["table"].astype(int)
data["table"] .replace(to_replace=range(53,59), value=8,inplace=True)
data["table"] .replace(to_replace=range(52,61), value=7,inplace=True)
data["table"] .replace(to_replace=range(51,64), value=6,inplace=True)
data["table"] .replace(to_replace=range(50,65), value=5,inplace=True)
data["table"] .replace(to_replace=range(50,68), value=4,inplace=True)
data["table"] .replace(to_replace=range(49,71), value=3,inplace=True)
data["table"] .replace(to_replace=range(48,73), value=2,inplace=True)
data["table"] .replace(to_replace=range(20,100), value=1,inplace=True)
data
```

## **Check data is ready**

```python
data.info()
```
